Analysing the variables

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(data.TARGET_deathRate, bins=30, color='#58C', edgecolor='k')
plt.xlabel('Death Rate')
plt.ylabel('Frequency')
plt.title('Histogram of Death Rate')
plt.show()
```

**Interpretations**
- vals are grouped bins along x-axis. Height represent no. of val fall in that bin.

We can answer the following questions :
- where is the center of data
- what is the spread of data range
- what is the shape of data? Like symmetric, skewed, uniform, bimodal etc.
