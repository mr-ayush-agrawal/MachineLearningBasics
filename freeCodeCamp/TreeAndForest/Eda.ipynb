```python
Train = data[data.Date.dt.year < 2015]
Val = data[data.Date.dt.year == 2015]
Test = data[data.Date.dt.year > 2015]

Train.shape, Val.shape, Test.shape
```

```python
Train.info()
```

```python
inp_cols = data.columns.tolist()[1:-1]
tar_col = data.columns.to_list()[-1]

inp_cols, tar_col
```

```python
x_train = Train[inp_cols]
y_train = Train[tar_col]

x_train.shape, y_train.shape
```

```python
num_cols = x_train.select_dtypes(include=np.number).columns.tolist()
cat_cols = x_train.select_dtypes('object').columns.tolist()
num_cols, cat_cols
```

```python
data[cat_cols].isna().sum()
```

```python
plt.figure(figsize=(14,8))
sns.heatmap(data[num_cols].corr(), center=0, cmap='Greens', annot=True, fmt='.2f')
```

### Imputing the values

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy= 'mean').fit(data[num_cols])
```

```python
x_train[num_cols] = imputer.transform(x_train[num_cols])
Test[num_cols] = imputer.transform(Test[num_cols])
Val[num_cols] = imputer.transform(Val[num_cols])

data[num_cols] = imputer.transform(data[num_cols])
```

```python
x_train.isna().sum()
```

```python
x_train[cat_cols].nunique()
```

### Scaling

```python
from sklearn.preprocessing import Normalizer
scaler = Normalizer().fit(data[num_cols])
```

```python
x_train[num_cols] = scaler.transform(x_train[num_cols])
Val[num_cols] = scaler.transform(Val[num_cols])
Test[num_cols] = scaler.transform(Test[num_cols])
```

```python
x_train.describe().T
```

### Filling categoriocal features

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore').fit(data[cat_cols])
```

```python
encoded_cols = list(encoder.get_feature_names_out(cat_cols))
len(encoded_cols)
```

```python
x_train[encoded_cols] = encoder.transform(x_train[cat_cols])
Val[encoded_cols] = encoder.transform(Val[cat_cols])
Test[encoded_cols] = encoder.transform(Test[cat_cols])
x_train.shape, Val.shape, Test.shape
```

```python
for col in Val.columns.tolist():
    if(col not in x_train.columns.tolist()):
        print(col)
```

```python
x_train = x_train[num_cols +  encoded_cols]
x_val = Val[num_cols + encoded_cols]
x_test = Test[num_cols + encoded_cols]

x_test.shape, x_train.shape, x_val.shape
```

```python
y_test = Test[tar_col]
y_val = Val[tar_col]

y_test.shape, y_val.shape
```

### Reducing Overfitting
Overfitting can be reduced by providing some of the hyperparamerter. to the model.

```python
??DecisionTreeClassifier
```

Using the `max_depth` parameter

```python
model = DecisionTreeClassifier(max_depth=3).fit(x_train, y_train)
train_pre = model.predict(x_train)
val_pre = model.predict(x_val)

print(accuracy_score(y_train, train_pre))
print(accuracy_score(y_val, val_pre))
# model.score(x_train, y_train)
```

There is improvement in the model

Running the same in the loop to findout the best fit

```python
depth = []
train_score = []
val_score = []

for i in range (1,16):
    print(i, end="\t")
    depth.append(i)
    model = DecisionTreeClassifier(max_depth=i).fit(x_train, y_train)
    train_score.append(model.score(x_train, y_train))
    val_score.append(model.score(x_val, y_val))

df = pd.DataFrame({'Depth': depth,
                   'Train Score': train_score, 
                   'Validation Score': val_score})
df
```

```python
plt.plot(df['Depth'], 1- df['Validation Score'], label = 'Validation Error')
plt.plot(df['Depth'], 1- df['Train Score'], label = 'Triani Error')
plt.legend()
plt.xlabel('Depth')
plt.ylabel('Error')
```

```python
df.sort_values('Validation Score', ascending=False)
```

From Here we can say thay best accuracy is at max_depth = 9,7

```python
model = DecisionTreeClassifier(max_depth=7).fit(x_train, y_train)
importance_df = pd.DataFrame({
    'feature': x_train.columns.tolist(),
    'importance' : model.feature_importances_
})
importance_df.sort_values('importance', ascending=False, inplace=True)
```

Using the hyperParam for improving the model `max_leaf_nodes`

```python
# Full depth but restricted Nodes

model = DecisionTreeClassifier(max_leaf_nodes=128).fit(x_train, y_train)
model.score(x_train, y_train)
```

Here we are getting more accuracy

```python
plt.figure(figsize=(10,8))
plot_tree(model, feature_names=x_train.columns)
plt.show()
```

```python
model.tree_.max_depth
```

#### Finding the optimal max Nodes
In step of 10

```python
depth = []
train_score = []
val_score = []

for i in range (9,260,10):
    print(i, end="\t")
    depth.append(i)
    model = DecisionTreeClassifier(max_leaf_nodes=i).fit(x_train, y_train)
    train_score.append(model.score(x_train, y_train))
    val_score.append(model.score(x_val, y_val))

df = pd.DataFrame({'Depth': depth,
                   'Train Score': train_score, 
                   'Validation Score': val_score})
df
```
