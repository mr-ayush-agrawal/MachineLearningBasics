### Splitting the data set 

```python
from sklearn.model_selection import train_test_split
from random import randint
```

Keeping Random state a constant will produce the same split in the model every time. THis makes the task easy for tracking, else there will be changes in the samples produced

```python
Train, Test = train_test_split(raw_data, test_size = 0.2, random_state=randint(0, 1000))
Train, Validate = train_test_split(Train, test_size=0.25, random_state=randint(0,1000))
Train.shape, Validate.shape, Test.shape
```

Time Based Splitting

```python
raw_data['Year']= pd.to_datetime(raw_data.Date).dt.year
raw_data.Year.nunique()
```

```python
sns.countplot(x=raw_data.Year)
```

```python
Train = raw_data[raw_data.Year < 2015]
Validate = raw_data[raw_data.Year == 2015]
Test = raw_data[raw_data.Year > 2015]

Train.shape, Validate.shape, Test.shape
```

```python
Train.head(10)
```

Segregating input and output cols

```python
input_cols = list(Train.columns)[1:-2]
tar_cols = ['RainTomorrow']
```

```python
X_train = Train[input_cols]
Y_train = Train[tar_cols]
```

```python
num_cols = X_train.select_dtypes(include= np.number).columns.tolist()[:-1]
cat_cols = X_train.select_dtypes('object').columns.tolist()
num_cols, cat_cols
```

```python
X_train[num_cols].describe().T
```

```python
X_train[cat_cols].nunique()
```

### Imputating the missing values

We are using the simple imputer from sklearn to fill the values. Strategy we are using is the mean

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy= 'mean')
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(X_train[num_cols].corr(), cmap = 'PuOr', annot= True, fmt='.2f', center=0)
```

```python
imputer.fit(X_train[num_cols])
```

```python
for i in range(len(num_cols)):
    print(num_cols[i], " ---> ", list(imputer.statistics_)[i])
```

```python
X_train[num_cols] = imputer.transform(X_train[num_cols])
X_train[num_cols].isna().sum()
```

```python
X_train.sample(5)
```

There are 2 ways now for filling the values of validate and test set

```python
Test[num_cols] = imputer.transform(Test[num_cols])
Validate[num_cols] = imputer.transform(Validate[num_cols])
```

### Encoding Categorical Data

```python
raw_data[cat_cols].nunique()
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder= OneHotEncoder(sparse= False, handle_unknown='ignore')
```

```python
encoder.fit(raw_data[cat_cols])
encoder.categories_
```

```python
encoded_cols = list(encoder.get_feature_names_out())
encoded_cols
```

```python
X_train[encoded_cols] = encoder.transform(X_train[cat_cols])
Validate[encoded_cols] = encoder.transform(Validate[cat_cols])
Test[encoded_cols] = encoder.transform(Test[cat_cols])
```
