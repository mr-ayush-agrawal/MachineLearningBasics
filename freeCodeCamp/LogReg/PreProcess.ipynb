This data has a lot of null values need to clean the data.

We can't work with the data where our target column `rainTomorrow` is missing as we have to get the train the model according to that. Similary we we don't have the data for `rainToday` This will again cause the same Issue. So we are dropping the values where these values are `NULL`

```python
raw_data.dropna(subset=['RainToday', 'RainTomorrow'], inplace=True)
```

### Splitting the data set 

```python
from sklearn.model_selection import train_test_split
from random import randint
```

Keeping Random state a constant will produce the same split in the model every time. THis makes the task easy for tracking, else there will be changes in the samples produced

```python
Train, Test = train_test_split(raw_data, test_size = 0.2, random_state=randint(0, 1000))
Train, Validate = train_test_split(Train, test_size=0.25, random_state=randint(0,1000))
Train.shape, Validate.shape, Test.shape
```
